```python
from IPython.display import display, Math
import math

def check(x):
  if isinstance(x, Expression):
    return x
  return Expression(x)

def expr2latex(x):
  if isinstance(x, (int, float)):
    return str(x)
  if isinstance(x, str):
    return x
  if isinstance(x, list):
    if x[0] == "^":
      return expr2latex(x[1]) + "^" + expr2latex(x[2])
    if x[0] == "--":
      a = expr2latex(x[1])
      if isinstance(x[1], list):
        a = "\\left(" + a + "\\right)"
      return "-{" + a + "}"
    if x[0] == "+":
      return expr2latex(x[1]) + "+" + expr2latex(x[2])
    if x[0] == "-":
      return expr2latex(x[1]) + "-" + expr2latex(x[2])
    if x[0] == "*":
      left = expr2latex(x[1])
      if isinstance(x[1], list) and x[1][0] == "+":
        left = "\\left(" + left + "\\right)"
      right = expr2latex(x[2])
      if isinstance(x[2], list) and x[2][0] == "+":
        right = "\\left(" + right + "\\right)"
      return left + "\\cdot " + right
    if x[0] == "/":
      return "\\dfrac{" + expr2latex(x[1]) + "}{" + expr2latex(x[2]) + "}"
    if x[0] == "arcsin":
      return "\\arcsin{" + expr2latex(x[1]) + "}"
    if x[0] == "arccos":
      return "\\arccos{" + expr2latex(x[1]) + "}"
    if x[0] == "arctg":
      return "\\arctg{" + expr2latex(x[1]) + "}"
    if x[0] == "arcctg":
      return "\\arcctg{" + expr2latex(x[1]) + "}"
    if x[0] == "sqrt":
      return "\\sqrt{" + expr2latex(x[1]) + "}"
    if x[0] == "ln":
      return "\\ln{" + expr2latex(x[1]) + "}"
    if x[0] == "lg":
      return "\\lg{" + expr2latex(x[1]) + "}"
    if x[0] == "sin":
      a = expr2latex(x[1])
      if isinstance(x[1], list):
        a = "\\left(" + a + "\\right)"
      return "\\sin{" + a + "}"
    if x[0] == "cos":
      a = expr2latex(x[1])
      if isinstance(x[1], list):
        a = "\\left(" + a + "\\right)"
      return "\\cos{" + a + "}"
    if x[0] == "tg":
      return "\\tg{" + expr2latex(x[1]) + "}"
    if x[0] == "ctg":
      return "\\ctg{" + expr2latex(x[1]) + "}"
    if x[0] == "sh":
      return "\\sh{" + expr2latex(x[1]) + "}"
    if x[0] == "ch":
      return "\\ch{" + expr2latex(x[1]) + "}"
    if x[0] == "th":
      return "\\th{" + expr2latex(x[1]) + "}"
    if x[0] == "cth":
      return "\\cth{" + expr2latex(x[1]) + "}"
    if x[0] == "arcsh":
      return "\\arcsinh{" + expr2latex(x[1]) + "}"
    if x[0] == "arcch":
      return "\\arccosh{" + expr2latex(x[1]) + "}"
    if x[0] == "arcth":
      return "\\arctanh{" + expr2latex(x[1]) + "}"
    if x[0] == "arccth":
      return "\\arcctanh{" + expr2latex(x[1]) + "}"


def sqrt(x):
  x = check(x)
  t = x.f
  r = fsqrt(t)
  return Expression(r)

def ln(x):
  x = check(x)
  t = x.f
  r = fln(t)
  return Expression(r)

def lg(x):
  x = check(x)
  t = x.f
  r = flg(t)
  return Expression(r)

def sin(x):
  return Expression(fsin(check(x).f))

def cos(x):
  return Expression(fcos(check(x).f))

def tg(x):
  x = check(x)
  t = x.f
  r = ftg(t)
  return Expression(r)

def ctg(x):
  x = check(x)
  t = x.f
  r = fctg(t)
  return Expression(r)

def arcsin(x):
  x = check(x)
  t = x.f
  r = farcsin(t)
  return Expression(r)

def arccos(x):
  x = check(x)
  t = x.f
  r = farccos(t)
  return Expression(r)

def arctg(x):
  x = check(x)
  t = x.f
  r = farctg(t)
  return Expression(r)

def arcctg(x):
  x = check(x)
  t = x.f
  r = farcctg(t)
  return Expression(r)

def sh(x):
  x = check(x)
  t = x.f
  r = fsh(t)
  return Expression(r)

def ch(x):
  x = check(x)
  t = x.f
  r = fch(t)
  return Expression(r)

def th(x):
  x = check(x)
  t = x.f
  r = fth(t)
  return Expression(r)

def cth(x):
  x = check(x)
  t = x.f
  r = fcth(t)
  return Expression(r)

def arcsh(x):
  x = check(x)
  t = x.f
  r = farccth(t)
  return Expression(r)

def arcch(x):
  x = check(x)
  t = x.f
  r = farccth(t)
  return Expression(r)

def arcth(x):
  x = check(x)
  t = x.f
  r = farccth(t)
  return Expression(r)

def arccth(x):
  x = check(x)
  t = x.f
  r = farccth(t)
  return Expression(r)

def farcsh(x):
  return ['arcsh',x]

def farcch(x):
  return ['arcch',x]
  
def farcth(x):
  return ['arcth',x]
  
def farccth(x):
  return ['arccth',x]
  
def fsh(x):
  return ['sh',x]

def fch(x):
  return ['ch',x]

def fth(x):
  return ['th',x]

def fcth(x):
  return ['cth',x]

def fsin(x):
  return ['sin',x]

def fcos(x):
  return ['cos',x]

def ftg(x):
  return ['tg',x]

def fctg(x):
  return ['ctg',x]

def farcsin(x):
  return ['arcsin',x]

def farccos(x):
  return ['arccos',x]

def farctg(x):
  return ['arctg',x]

def farcctg(x):
  return ['arcctg',x]

def fln(x):
  return ['ln',x]

def flg(x):
  return ['lg',x]

def fsqrt(x):
  return ['sqrt', x]

def add(x, y):
  return ['+', x, y]

def mul(x, y):
  return ['*', x, y]

def sub(x, y):
  return ['-', x, y]

def div(x, y):
  return ["/", x, y]

def pow(x, y):
  return ["^", x, y]
def neg(x):
  return ["--", x]

def substitute(f, var, val):
  val = check(val).f
  if f == var:
    return val
  elif isinstance(f, list):
    y = [f[0]]
    for i in range(1, len(f)):
      t = substitute(f[i], var, val)
      y.append(t)
    return y
  else:
    return f

def evaluate(f):
  if isinstance(f, (int, float)):
    return f
  elif isinstance(f, list):
    if f[0] == "+": 
      return evaluate(f[1]) + evaluate(f[2])
    elif f[0] == "*":
      return evaluate(f[1]) * evaluate(f[2])
    elif f[0] == "sqrt":
      return math.sqrt(evaluate(f[1]))
    elif f[0] == "sin":
      return math.sin(evaluate(f[1]))
    elif f[0] == "cos":
      return math.cos(evaluate(f[1]))
  return None

class Expression:
  def __init__(self, f):
    self.f = f
  def __add__(self, x):
    x = check(x)
    return Expression(add(self.f, x.f))
  def __sub__(self, x):
    x = check(x)
    return Expression(sub(self.f, x.f))
  def __mul__(self, x):
    x = check(x)
    return Expression(mul(self.f, x.f))
  def __truediv__(self, x):
    x = check(x)
    return Expression(div(self.f, x.f))
  def __pow__(self, x):
    x = check(x)
    return Expression(pow(self.f, x.f))
  def __radd__(self, x):
    x = check(x)
    return Expression(add(x.f, self.f))
  def __rsub__(self, x):
    x = check(x)
    return Expression(sub(x.f, self.f))
  def __rmul__(self, x):
    x = check(x)
    return Expression(mul(x.f, self.f))
  def __rtruediv__(self, x):
    x = check(x)
    return Expression(div(x.f, self.f))
  def __rpow__(self, x):
    x = check(x)
    return Expression(pow(x.f, self.f))
  def __neg__(self):
    return Expression(neg(self.f))
  def subs(self, var, val):
    return Expression(substitute(self.f, var, val))
  def evalf(self):
    return evaluate(self.f)
  def __str__(self):
    return expr2latex(self.f)
  def show(self):
    display(Math(expr2latex(self.f)))
  def diff(self, var):
    var = check(var).f
    return Expression(differentiate(self.f, var))
  def simp(self):
    return Expression(simplify(self.f))

def evaluate(f):
  if isinstance(f, (int, float)):
    return str(f)
  if isinstance(f, str):
    return f
  elif isinstance(f, list):
    if f[0] == "+": 
      return evaluate(f[1]) + evaluate(f[2])
    elif f[0] == "*":
      return evaluate(f[1]) * evaluate(f[2])
    elif f[0] == "sqrt":
      return math.sqrt(evaluate(f[1]))
    elif f[0] == "sin":
      return math.sin(evaluate(f[1]))
    elif f[0] == "cos":
      return math.cos(evaluate(f[1]))
    elif f[0] == "tg":
      return math.tg(evaluate(f[1]))
    elif f[0] == "ctg":
      return math.ctg(evaluate(f[1]))
    elif f[0] == "arcsin":
      return math.arcsin(evaluate(f[1]))
    elif f[0] == "arccos":
      return math.arccos(evaluate(f[1]))
    elif f[0] == "arctg":
      return math.arctg(evaluate(f[1]))
    elif f[0] == "arcctg":
      return math.arcctg(evaluate(f[1]))
  return None

def symbols(vars):
  return map(Expression, vars.split())
x, y = symbols("x y")  

def differentiate(f, x):
  if isinstance(f, int):
    return 0
  if f == x:
    return 1
  if isinstance(f, str):
    return 0
  if isinstance(f, list):
    if f[0] == "+":
      u, v = f[1], f[2]
      u1, v1 = differentiate(u, x), differentiate(v, x)
      return add(u1, v1)
    if f[0] == "-":
      u, v = f[1], f[2]
      u1, v1 = differentiate(u, x), differentiate(v, x)
      return sub(u1, v1)      
    if f[0] == "*":
      u, v = f[1], f[2]
      u1, v1 = differentiate(u, x), differentiate(v, x)
      return add(mul(u1, v), mul(u, v1))   
    if f[0] == "/":
      u, v = f[1], f[2]
      u1, v1 = differentiate(u, x), differentiate(v, x)
      return div(sub(mul(u1, v), mul(u, v1)), pow(v, 2))   
    if f[0] == "^":
      u, v = f[1], f[2]
      u1, v1 = differentiate(u, x), differentiate(v, x)
      p1, p2 = pow(u, v), pow(u, sub(v, 1))
      s1 = mul(p1, mul(v1, fln(u)))
      s2 = mul(v, mul(p2, u1))
      return add(s1, s2) 
    if f[0] == "--":
      u = f[1]
      u1 = differentiate(u, x)
      return neg(u1)
    if f[0] == "ln":
      u = f[1]
      u1 = differentiate(u, x)
      return div(u1, u)
    if f[0] == "lg":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = mul(u,fln(10))
      return div(u1, u2)
    if f[0] == "sin":
      u = f[1]
      u1 = differentiate(u, x)
      return mul(fcos(u), u1)
    if f[0] == "cos":
      u = f[1]
      u1 = differentiate(u, x)
      return mul(neg(fsin(u)), u1)
    if f[0] == "sqrt":
      u = f[1]
      u1 = differentiate(u, x)
      return div(u1, mul(2, fsqrt(u)))
    if f[0] == "tg":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = mul(fcos(u),fcos(u))
      return div(u1,u2)
    if f[0] == "ctg":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = mul(fsin(u),fsin(u))
      return div(neg(u1),u2)
    if f[0] == "sh":
      u = f[1]
      u1 = differentiate(u, x)
      return mul(u1,fch(u))
    if f[0] == "ch":
      u = f[1]
      u1 = differentiate(u, x)
      return mul(u1,fsh(u))
    if f[0] == "th":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = mul(fch(u),fch(u))
      return div(u1,u2)
    if f[0] == "cth":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = mul(fsh(u),fsh(u))
      return div(neg(u1),u2)
    if f[0] == "arcsin":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = fsqrt(sub(1,mul(u,u)))
      return div(u1,u2)
    if f[0] == "arccos":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = fsqrt(sub(1,mul(u,u)))
      return div(neg(u1),u2)
    if f[0] == "arctg":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = add(1,mul(u,u))
      return div(u1,u2)
    if f[0] == "arcctg":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = add(1,mul(u,u))
      return div(neg(u1),u2)
    if f[0] == "arcsh":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = fsqrt(add(mul(u,u),1))
      return div(u1,u2)
    if f[0] == "arcch":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = fsqrt(sub(mul(u,u),1))
      return div(u1,u2)
    if f[0] == "arcth":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = sub(1,mul(u,u))
      return div(neg(u1),u2)
    if f[0] == "arccth":
      u = f[1]
      u1 = differentiate(u, x)
      u2 = sub(1,mul(u,u))
      return div(u1,u2)
  
def simplify(f):
  if isinstance(f, list):
    r = [f[0]]
    for x in f[1:]:
      r.append(simplify(x))
    if r[0] == "+":
      if isinstance(r[1], int) and isinstance(r[2], int):
        return r[1] + r[2] 
      if r[1] == 0: 
        return r[2]
      if r[2] == 0:
        return r[1]
      if r[1] == r[2]:
        return mul(2, r[1])
      if isinstance(r[2], list) and r[2][0] == "--":
        return simplify(sub(r[1], r[2][1]))
    if r[0] == "-":
      if isinstance(r[1], int) and isinstance(r[2], int):
        return r[1] - r[2]
      if r[1] == 0: 
        return neg(r[2])
      if r[2] == 0:
        return r[1]
      if r[1] == r[2]:
        return 0
      if isinstance(r[2], list) and r[2][0] == "--":
        return simplify(add(r[1], r[2][1]))
    if r[0] == "*":
      if isinstance(r[1], int) and isinstance(r[2], int):
        return r[1] * r[2] 
      if r[1] == 0 or r[2] == 0: 
        return 0
      if r[1] == 1:
        return r[2]
      if r[2] == 1:
        return r[1]
      if r[1] == r[2]:
        return pow(r[1], 2)
    if r[0] == "^":
      if isinstance(r[1], int) and isinstance(r[2], int) and r[2] > 0:
        return r[1] ** r[2] 
      if r[2] == 1: 
        return r[1]
      if r[2] == 0: 
        return 1
    if r[0] == "/":
      if r[2] == 1: 
        return r[1]
      if r[1] == 0:
        return 0
      if isinstance(r[1], int) and isinstance(r[2], int):
        if r[1] > r[2]:
          a = r[1]
          b=r[2]
        else:
          a = r[2]
          b = r[1]
        while a != 0 and b != 0:
          if a >= b:
              a -= b
          else:
              b -= a
        if a!=0:
          return div(r[1]/a,r[2]/a)
        else:
          return div(r[1]/b,r[2]/b)
    if r[0] == "sqrt":
      if isinstance(r[1], int):
        a = int(math.sqrt(r[1]))
        if a * a == r[1]:
          return a
        return r 
    return r
  return f
```
