```python
import math
from IPython.display import display, Math

def check(x):
  if isinstance(x, Expression):
    return x
  return Expression(x)

def expr2latex(x):
  if isinstance(x, (int, float)):
    return str(x)
  if isinstance(x, str):
    return x
  if isinstance(x, list):
    if x[0] == "**":
      return expr2latex(x[1]) + "^" + expr2latex(x[2])
    if x[0] == "+":
      return expr2latex(x[1]) + "+" + expr2latex(x[2])
    if x[0] == "-":
      return expr2latex(x[1]) + "-" + expr2latex(x[2])
    if x[0] == "*":
      left = expr2latex(x[1])
      if isinstance(x[1], list) and x[1][0] == "+":
        left = "(" + left + ")"
      right = expr2latex(x[2])
      if isinstance(x[2], list) and x[2][0] == "+":
        right = "(" + right + ")"
      return left + "\\cdot " + right
    if x[0] == "/":
      return "\\dfrac{" + expr2latex(x[1]) + "}{" + expr2latex(x[2]) + "}"
    if x[0] == "arcsin":
      return "\\arcsin{" + expr2latex(x[1]) + "}"
    if x[0] == "arccos":
      return "\\arccos{" + expr2latex(x[1]) + "}"
    if x[0] == "arctg":
      return "\\arctg{" + expr2latex(x[1]) + "}"
    if x[0] == "arcctg":
      return "\\arcctg{" + expr2latex(x[1]) + "}"
    if x[0] == "sqrt":
      return "\\sqrt{" + expr2latex(x[1]) + "}"
    if x[0] == "ln":
      return "\\ln{" + expr2latex(x[1]) + "}"
    if x[0] == "lg":
      return "\\lg{" + expr2latex(x[1]) + "}"
    if x[0] == "sin":
      return "\\sin{" + expr2latex(x[1]) + "}"
    if x[0] == "cos":
      return "\\cos{" + expr2latex(x[1]) + "}"
    if x[0] == "tg":
      return "\\tg{" + expr2latex(x[1]) + "}"
    if x[0] == "ctg":
      return "\\ctg{" + expr2latex(x[1]) + "}"
    if x[0] == "sh":
      return "\\sh{" + expr2latex(x[1]) + "}"
    if x[0] == "ch":
      return "\\ch{" + expr2latex(x[1]) + "}"
    if x[0] == "th":
      return "\\th{" + expr2latex(x[1]) + "}"
    if x[0] == "cth":
      return "\\cth{" + expr2latex(x[1]) + "}"
    if x[0] == "arcsh":
      return "\\arcsinh{" + expr2latex(x[1]) + "}"
    if x[0] == "arcch":
      return "\\arccosh{" + expr2latex(x[1]) + "}"
    if x[0] == "arcth":
      return "\\arctanh{" + expr2latex(x[1]) + "}"
    if x[0] == "arccth":
      return "\\arcctanh{" + expr2latex(x[1]) + "}"


def sqrt(x):
  x = check(x)
  t = x.f
  r = fsqrt(t)
  return Expression(r)

def ln(x):
  x = check(x)
  t = x.f
  r = fln(t)
  return Expression(r)

def lg(x):
  x = check(x)
  t = x.f
  r = flg(t)
  return Expression(r)

def sin(x):
  x = check(x)
  t = x.f
  r = fsin(t)
  return Expression(r)

def cos(x):
  x = check(x)
  t = x.f
  r = fcos(t)
  return Expression(r)

def tg(x):
  x = check(x)
  t = x.f
  r = ftg(t)
  return Expression(r)

def ctg(x):
  x = check(x)
  t = x.f
  r = fctg(t)
  return Expression(r)

def arcsin(x):
  x = check(x)
  t = x.f
  r = farcsin(t)
  return Expression(r)

def arccos(x):
  x = check(x)
  t = x.f
  r = farccos(t)
  return Expression(r)

def arctg(x):
  x = check(x)
  t = x.f
  r = farctg(t)
  return Expression(r)

def arcctg(x):
  x = check(x)
  t = x.f
  r = farcctg(t)
  return Expression(r)

def sh(x):
  x = check(x)
  t = x.f
  r = fsh(t)
  return Expression(r)

def ch(x):
  x = check(x)
  t = x.f
  r = fch(t)
  return Expression(r)

def th(x):
  x = check(x)
  t = x.f
  r = fth(t)
  return Expression(r)

def cth(x):
  x = check(x)
  t = x.f
  r = fcth(t)
  return Expression(r)

def arcsh(x):
  x = check(x)
  t = x.f
  r = farccth(t)
  return Expression(r)

def arcch(x):
  x = check(x)
  t = x.f
  r = farccth(t)
  return Expression(r)

def arcth(x):
  x = check(x)
  t = x.f
  r = farccth(t)
  return Expression(r)

def arccth(x):
  x = check(x)
  t = x.f
  r = farccth(t)
  return Expression(r)

def farcsh(x):
  return ['arcsh',x]

def farcch(x):
  return ['arcch',x]
  
def farcth(x):
  return ['arcth',x]
  
def farccth(x):
  return ['arccth',x]
  
def fsh(x):
  return ['sh',x]

def fch(x):
  return ['ch',x]

def fth(x):
  return ['th',x]

def fcth(x):
  return ['cth',x]

def fsin(x):
  return ['sin',x]

def fcos(x):
  return ['cos',x]

def ftg(x):
  return ['tg',x]

def fctg(x):
  return ['ctg',x]

def farcsin(x):
  return ['arcsin',x]

def farccos(x):
  return ['arccos',x]

def farctg(x):
  return ['arctg',x]

def farcctg(x):
  return ['arcctg',x]

def fln(x):
  return ['ln',x]

def flg(x):
  return ['lg',x]

def fsqrt(x):
  return ['sqrt', x]

def add(x, y):
  return ['+', x, y]

def mul(x, y):
  return ['*', x, y]

def sub(x, y):
  return ['-', x, y]

def div(x, y):
  return ["/", x, y]

def pow(x, y):
  return ["**", x, y]

def substitute(f, var, val):
  val = check(val).f
  if f == var:
    return val
  elif isinstance(f, list):
    y = [f[0]]
    for i in range(1, len(f)):
      t = substitute(f[i], var, val)
      y.append(t)
    return y
  else:
    return f

def evaluate(f):
  if isinstance(f, (int, float)):
    return f
  elif isinstance(f, list):
    if f[0] == "+": 
      return evaluate(f[1]) + evaluate(f[2])
    elif f[0] == "*":
      return evaluate(f[1]) * evaluate(f[2])
    elif f[0] == "sqrt":
      return math.sqrt(evaluate(f[1]))
    elif f[0] == "sin":
      return math.sin(evaluate(f[1]))
    elif f[0] == "cos":
      return math.cos(evaluate(f[1]))
  return None

class Expression:
  def __init__(self, f):
    self.f = f
  def __add__(self, x):
    x = check(x)
    return Expression(add(self.f, x.f))
  def __pow__(self, x):
    x = check(x)
    return Expression(pow(self.f, x.f))
  def __mul__(self, x):
    x = check(x)
    return Expression(mul(self.f, x.f))
  def __sub__(self, x):
    x = check(x)
    return Expression(sub(self.f, x.f))
  def __truediv__(self, x):
    x = check(x)
    return Expression(div(self.f, x.f))
  def subs(self, var, val):
    return Expression(substitute(self.f, var, val))
  def evalf(self):
    return evaluate(self.f)
  def __str__(self):
    return expr2latex(self.f)
  def show(self):
    display(Math(expr2latex(self.f)))

def evaluate(f):
  if isinstance(f, (int, float)):
    return f
  elif isinstance(f, list):
    if f[0] == "+": 
      return evaluate(f[1]) + evaluate(f[2])
    elif f[0] == "*":
      return evaluate(f[1]) * evaluate(f[2])
    elif f[0] == "sqrt":
      return math.sqrt(evaluate(f[1]))
    elif f[0] == "sin":
      return math.sin(evaluate(f[1]))
    elif f[0] == "cos":
      return math.cos(evaluate(f[1]))
    elif f[0] == "tg":
      return math.tg(evaluate(f[1]))
    elif f[0] == "ctg":
      return math.ctg(evaluate(f[1]))
    elif f[0] == "arcsin":
      return math.arcsin(evaluate(f[1]))
    elif f[0] == "arccos":
      return math.arccos(evaluate(f[1]))
    elif f[0] == "arctg":
      return math.arctg(evaluate(f[1]))
    elif f[0] == "arcctg":
      return math.arcctg(evaluate(f[1]))
  return None
```
