```python
from sympy import *

t, m, k, n= symbols('t m k n',real=True)
v = Function('v')
v(t)


func = Eq(Derivative(v(t))*m,-k*v(t))
ic = {v(0):10} 
a = dsolve(func, v(t), ics = ic)
ic = {v(20):5} 


sh = a.subs(v(t), 5).subs(t,20).simplify().subs(k/m, n)
n = solve(sh, n)
a.simplify().subs(k/m,n[0]).subs(t,60)

```

```python
from sympy import *
from sympy.plotting import *

t, m, k, n, P= symbols('t m k n P',real=True)
v = Function('v')
h = Function('h')
v(t)

P = 1.2*5/6*9.8
func = Eq(Derivative(v(t))*1,P-0.001*v(t))
ic = {v(0):1}

a = dsolve(func, v(t), ics = ic)

a.subs(t, 0)

integrate(a).subs(t,0)
a.subs(t,100)

p1 = plot(a.rhs, (t,0,200))



h = integrate(a).rhs - integrate(a).subs(t,0).rhs

p2 = plot(h, (t,0,200))

a.subs(t,50)

h
```
